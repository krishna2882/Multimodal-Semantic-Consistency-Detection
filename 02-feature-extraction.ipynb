```python
import os
import re
import emoji
import numpy as np
import pandas as pd
from PIL import Image
from tqdm import tqdm
from collections import Counter

import torch
from transformers import CLIPProcessor, CLIPModel
```

```python
df = pd.read_csv("/kaggle/input/notebooks/krishnaiiith/01-dataset-preparation/labeled.csv")
print("Dataset size:", len(df))
df.head()
```

```python
def clean_text(text):

    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"@\w+", "", text)
    text = emoji.demojize(text)

    return text.lower().strip()


def extract_hashtags(text):

    return re.findall(r"#\w+", text)


def segment_hashtag(tag):

    tag = tag.replace("#", "")
    tag = re.sub(r'([a-z])([A-Z])', r'\1 \2', tag)
    tag = re.sub(r'(\d+)', r' \1 ', tag)

    return tag.lower()
```

```python
text_inputs = []
hashtag_inputs = []
image_paths = []
all_hashtags = []

for _, row in df.iterrows():

    clean = clean_text(row["text"])

    hashtags = extract_hashtags(row["text"])
    segmented = [segment_hashtag(tag) for tag in hashtags]

    hashtag_text = " ".join(segmented) if segmented else "none"

    text_inputs.append(clean)
    hashtag_inputs.append(hashtag_text)
    image_paths.append(row["image_path"])
    all_hashtags.append(hashtags)
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"

model = CLIPModel.from_pretrained(
    "openai/clip-vit-base-patch32"
).to(device)

processor = CLIPProcessor.from_pretrained(
    "openai/clip-vit-base-patch32"
)

model.eval()
```

```python
def get_text_embeddings(text_list, batch_size=32):

    embeddings = []

    for i in tqdm(range(0, len(text_list), batch_size)):

        batch = text_list[i:i+batch_size]

        inputs = processor(
            text=batch,
            padding=True,
            truncation=True,
            return_tensors="pt"
        )

        input_ids = inputs["input_ids"].to(device)
        attention_mask = inputs["attention_mask"].to(device)

        with torch.no_grad():

            outputs = model.text_model(
                input_ids=input_ids,
                attention_mask=attention_mask
            )

            pooled = outputs.pooler_output
            features = model.text_projection(pooled)

        embeddings.append(features.cpu())

    return torch.cat(embeddings).numpy()
```

```python
def get_image_embeddings(image_paths, batch_size=32):

    embeddings = []

    for i in tqdm(range(0, len(image_paths), batch_size)):

        batch_paths = image_paths[i:i+batch_size]
        images = []

        for path in batch_paths:

            try:
                img = Image.open(path).convert("RGB")
            except:
                img = Image.new("RGB",(224,224))

            images.append(img)

        inputs = processor(images=images, return_tensors="pt")

        pixel_values = inputs["pixel_values"].to(device)

        with torch.no_grad():

            outputs = model.vision_model(pixel_values=pixel_values)
            pooled = outputs.pooler_output
            features = model.visual_projection(pooled)

        embeddings.append(features.cpu())

    return torch.cat(embeddings).numpy()
```

```python
import re

def extract_hashtags(text):
    tags = re.findall(r"#\w+", text)
    if len(tags) == 0:
        return ""
    return " ".join(tags)

df["hashtags"] = df["text"].apply(extract_hashtags)
```

```python
def get_hashtag_embeddings(hashtags, batch_size=32):

    embeddings = []

    for i in tqdm(range(0, len(hashtags), batch_size)):

        batch = hashtags[i:i+batch_size]

        inputs = processor(
            text=batch,
            padding=True,
            truncation=True,
            return_tensors="pt"
        ).to(device)

        with torch.no_grad():

            outputs = model.get_text_features(
                input_ids=inputs["input_ids"],
                attention_mask=inputs["attention_mask"]
            )

        embeddings.append(outputs.pooler_output.cpu())

    embeddings = torch.cat(embeddings, dim=0)

    return embeddings.numpy()
```

```python
hashtag_inputs = df["hashtags"].tolist()
```

```python
print("Encoding text...")
text_emb = get_text_embeddings(text_inputs)

print("Encoding hashtags...")
hash_emb = get_hashtag_embeddings(hashtag_inputs)

print("Encoding images...")
img_emb = get_image_embeddings(image_paths)
```

```python
from collections import Counter
import numpy as np

# Flatten hashtag list
flat_tags = [tag for tags in all_hashtags for tag in tags]

# Count hashtag frequency in dataset
freq_dict = Counter(flat_tags)

# Number of hashtags in each tweet
hashtag_count = np.array([
    len(tags) for tags in all_hashtags
]).reshape(-1,1)

# Frequency score of hashtags used in tweet
hashtag_freq = np.array([
    sum(freq_dict[t] for t in tags) if tags else 0
    for tags in all_hashtags
]).reshape(-1,1)
```

```python
np.save("text_emb.npy", text_emb)
np.save("hash_emb.npy", hash_emb)
np.save("img_emb.npy", img_emb)

np.save("hashtag_count.npy", hashtag_count)
np.save("hashtag_freq.npy", hashtag_freq)

print("Feature extraction complete.")
```

```python
import numpy as np

# Normalize embeddings (important for cosine similarity)
text_norm = text_emb / np.linalg.norm(text_emb, axis=1, keepdims=True)
img_norm  = img_emb  / np.linalg.norm(img_emb,  axis=1, keepdims=True)
hash_norm = hash_emb / np.linalg.norm(hash_emb, axis=1, keepdims=True)

# Cosine similarities
text_img_sim  = np.sum(text_norm * img_norm, axis=1, keepdims=True)
text_hash_sim = np.sum(text_norm * hash_norm, axis=1, keepdims=True)
img_hash_sim  = np.sum(img_norm  * hash_norm, axis=1, keepdims=True)

# Final semantic feature matrix
semantic_features = np.concatenate([
    text_img_sim,
    text_hash_sim,
    img_hash_sim
], axis=1)

print("Semantic feature shape:", semantic_features.shape)
```
